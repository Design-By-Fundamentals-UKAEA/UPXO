```python
from upxo.ggrowth.mcgs import mcgs
import matplotlib.pyplot as plt
import numpy as np
```

```python
pxtal = mcgs(study='independent', input_dashboard='input_dashboard_for_testing_50x50_alg202.xls')
pxtal.simulate()
pxtal.detect_grains()
tslice = 10
pxtal.gs[tslice].char_morph_2d(char_gb=True)
```

```python
GIDs, VALIND = pxtal.gs[tslice].get_gids_in_params_bounds(mpnames=['aspect_ratio', 'area'],
                                              fx_stats=[np.mean, np.mean],
                                              pdslh=[[50, 30], [50, 30]], plot_mprop=False
                                              )
pxtal.gs[tslice].map_scalar_to_lgi(GIDs['presence'], default_scalar=-1,
                              plot=True, throw_axis=True)
```
