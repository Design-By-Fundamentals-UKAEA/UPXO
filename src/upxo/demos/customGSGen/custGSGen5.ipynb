```python
from upxo.pxtalops.temporalPatcher import patcher_pixvox as PAT
from upxo.ggrowth.mcgs import mcgs
import matplotlib.pyplot as plt
```

```python
pxt = mcgs(input_dashboard = 'C:\\Development\\UPXO\\upxo_library\\src\\upxo\\demos\\customGSGen\\custGSGen1.xls')
pxt.simulate()
pxt.detect_grains(library = 'scikit-image')
```

```python
pxt.uigrid
```

```python
pat = PAT.by_lgi(base_lfi=pxt.gs[15].lfi, patch_lfi=pxt.gs[1].lfi, 
            ninstances=1, constraints=None, dim=2)
```

```python
constraints = {'xbound': [10, 90], 'ybound': [10, 90], 'cellPropagation': 'b2p'}
pat.patch(lfib=pat.lfib, lfip=pat.lfip, patchDomain='rectangular',
          remove_small_features=False, constraints=constraints)
plt.imshow(pat.lfi_patched[1], cmap='nipy_spectral')
```

```python
constraints = {'xcenter': 50, 'ycenter': 50, 'radius': 40, 'cellPropagation': 'b2p'}
pat.patch(lfib=pat.lfib, lfip=pat.lfip, patchDomain='circular',
          remove_small_features=False, constraints=constraints)
plt.imshow(pat.lfi_patched[1], cmap='nipy_spectral')
```
