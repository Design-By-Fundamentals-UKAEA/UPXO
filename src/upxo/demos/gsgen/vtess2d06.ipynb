```python
repr_prop={'area': {'mean': {'val': 20, 'dev': 2.5, },
                    'consider_boundary_grains': True } }
```

# Examples for generating GS

gtess2d offers various instance creation possibilities. All of them are presenyted in this notebook.
We will explore the following in this notebook:
- By intrinsically generating seed points.
    - Random points
    - Poisson disc sampling by Bridson's algorithm
    - Dart sampling based seed point
    - Regular grid
        - Triangular
        - Rectangular
        - Mixed
- By importing pre-existing seed points.
    - Through file
    - Function argument input

## Poisson disc sampling: Bridson's algorithm

For this we will need, 
- GS is created by generating random seed points (sp[0]=’random’) of a Poisson disc sampling (sp[1]=’pds’) using Bridson’s algorithm (s[2]=bridson1’) and UPXO seed point type (sp[3]=’veryhigh’).
- ‘veryhigh’ indicates creating instances having thin memory requirements and shorter instantiation times, but with lesser functionalities).
- “char_length” is the minimum allowable inter-seed point distance of the Poisson’s disc sampling and determines how dense and uniform the seed points would be. A greater number would generate less points within same domain area.
- “ntp” is the number of trial points to be generated inside the Poisson’s annulus of radii char_length[0] and 2*char_length[1].
- “ntrials” is the number of trials used in the creation of the 1st pxtal instance. The iterations will be done to ensure pxtal parameter agrees to user specifications in “repr_prop”.
- “k_char_length_inc” and “k_char_length_dec”. Factor to increase or decrease the characteristic length respectively during iterations. Value must be greater than 0.

```python
gset = gtess2d.from_seed_points(sp_input='gen', xbound=[0, 100],
         ybound = [0, 50], sp_distr='random', gr_tech='pds',
         smp_tech='bridson1', lean='veryhigh', char_length=[5],
         niter=10, ntrials=100, n_instances=4, repr_prop=repr_prop,
         k_char_length_inc=0.05, k_char_length_dec=0.05,)
gset.plot(1)
```

## Dart sampling

```python
gset = gtess2d.from_seed_points(sp_input='gen', xbound=[0, 100],
                 ybound = [0, 100], sp_distr='random', gr_tech='random',
                 smp_tech='dart', lean='veryhigh', char_length=[4.5],
                 niter=10, ntrials=100, n_instances=4, repr_prop=repr_prop,
                 k_char_length_inc=0.05, k_char_length_dec=0.05,)
gset.plot(4)
```
