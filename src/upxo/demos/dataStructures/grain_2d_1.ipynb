### Accessing a MCGS2D grain object

1. Access the grain dictionary inside the temporal slice using `pxt.gs[tslice].g`.
2. This has two fields, `s` and `grain`, the latter being the actual grain object and the former being the MC state value which the grainobject belongs to. Key is the grain number. **Note**: This is planned for optimzation by converting the dict into a list of grain objects. `pxt.gs[tslice].s_gid` and `pxt.gs[tslice].gid_s` will be used to store the gid - s mappings instead.

```python
# For example, get grain object with gid=1
gid = 1 
print(pxt.gs[tslice].g[gid]['s'])
pxt.gs[tslice].g[gid]['grain']
```

As you can see, the `print(pxt.gs[tslice].g[gid]['s'])` printed 1, meaning the grain #1 correspomnds to q=1 of MC states 1:Q. The second output is the UPXo grain2D object


#### Lets get all the grains now.

```python
all_grains = [pxt.gs[tslice].g[i]['grain'] for i in pxt.gs[tslice].g.keys()]
all_grains[:4]
```
