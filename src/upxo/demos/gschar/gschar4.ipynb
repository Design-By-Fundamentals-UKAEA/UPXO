# Surface - sub-surface relationships

**Objective**: To carry out preliminary studies on surface - sub-surface grain struvctrue morphological relationmships.

We will use the definition **sss_rel_morpho** and **sss_rel_morpho_multiple**, def attributes of the 3D MCGS temporal slice.

### We will first create the grain structure, characterize it and calculate morppholocal parameters

```python
from upxo.ggrowth.mcgs import mcgs
import numpy as np
input_dashboard='C:\\Development\\UPXO\\upxo_library\\src\\upxo\\demos\\gschar\\gschar3.xls'
pxt = mcgs(input_dashboard=input_dashboard)

pxt.simulate(verbose=False)
tslice = 12
gstslice = pxt.gs[tslice]
gstslice.char_morphology_of_grains()
gstslice.set_mprops()
```

***
### Surface to sub-surface morphological relationships for 3D RVE with single slice.
> Before we find the relationships we are after, we need to first create the grain voxel locations and the spatial bounds of grains. The former creates coordinate triplets and the latter creates slice data structures which representy the spatial boudns of grains, which will be internally used to calculate the surface to sub-urface relationships.

```python
gstslice.find_grain_voxel_locs_v1(gids=None, disp_msg=False, verbosity=100,
                saa=True, throw=False, dtype=np.int32, use_uint16=True)
gstslice.find_spatial_bounds_of_grains()
```

```python
gstslice.sss_rel_morpho(slice_plane='xy', loc=24, reset_lgi=True, slice_gschar_kernel_order=4,
            mprop_names_2d=['eqdia', 'arbbox', 'solidity'], mprop_names_3d=['eqdia', 'arbbox', 'solidity'], 
            ignore_border_grains_2d=True, ignore_border_grains_3d=True, reset_mprops=False,
            kdeplot=True, save_plot3d_grains=True, save_plot2d_grains=True)
```
