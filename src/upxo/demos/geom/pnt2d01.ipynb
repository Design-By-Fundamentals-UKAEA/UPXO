# UPXO 2D Point object

## Introduction

This notebook demonstrates the **UPXO 2D Point** module, which provides two core classes for working with 2D points:

- **`Point2d`**: A feature-rich 2D point class with extensive geometric operations, transformations, and interoperability with external libraries (Shapely, GMSH, PyVista).
- **`p2d_leanest`**: A minimal-footprint 2D point class optimized for memory efficiency and performance-critical applications.

**`Point2d`** offers comprehensive functionality including:
- Equality/inequality comparisons with multiple input types (points, lists, tuples, coordinate arrays)
- Distance calculations (squared and Euclidean)
- Geometric operations (translation, rotation, line intersection)
- Spatial queries (nearest neighbors, clustering, point-in-line tests)
- Multi-format export (NumPy arrays, Shapely geometries, mesh formats)

**`p2d_leanest`** provides basic coordinate storage with minimal overhead, suitable for large-scale point collections where memory is constrained.

Both classes support flexible input formats and can interoperate seamlessly, allowing you to choose the appropriate class based on your performance and functionality requirements.

# Demonstrations

### Imports and aliases
Load `Point2d` (feature-rich) and `p2d_leanest` (minimal footprint) so each example can contrast the full and lean variants. We also pull numpy/matplotlib for potential visualization.

```python
from upxo.geoEntities.point2d import Point2d as p2d
from upxo.geoEntities.point2d import p2d_leanest
import numpy as np
import matplotlib.pyplot as plt
from upxo.geoEntities.point2d import Point2d
```

### Equality via operator
Shows `==` comparisons between `Point2d` and `p2d_leanest`, plus list/tuple inputs, to clarify how the rich and lean classes normalize coordinates.

```python
print(p2d(3, 4) == p2d_leanest(3, 4))
print(p2d(3, 4) == [p2d_leanest(1, 4), p2d_leanest(3, 4)])
print(p2d(3, 4) == p2d(3, 4))
print(p2d(3, 4) == [p2d(1, 2), p2d(3, 4)])
print(p2d(3, 4) == (p2d(1, 4), p2d(3, 4)))

print(p2d(3, 4) == [[1, 2], [3, 4], [5, 6]])
print(p2d(3, 4) == [[3, 4]])
print(p2d(3, 4) == [[3], [4]])
```

### Equality via `.eq()`
Demonstrates the explicit `eq` helper on `Point2d` to compare against lean points, lists, tuples, and nested coordinate lists—useful when you want method semantics instead of `==`.

```python
print(p2d(3, 4).eq(p2d_leanest(3, 4)))
print(p2d(3, 4).eq([p2d_leanest(1, 4), p2d_leanest(3, 4)]))
print(p2d(3, 4).eq(p2d(3, 4)))
print(p2d(3, 4).eq([p2d(1, 2), p2d(3, 4)]))
print(p2d(3, 4).eq((p2d(1, 4), p2d(3, 4))))
print(p2d(3, 4).eq([[1, 2], [3, 4], [5, 6]]))
print(p2d(3, 4).eq([[3, 4]]))
print(p2d(3, 4).eq([[3], [4]]))
```

### Fast equality with `point_spec`
Exercises `eq_fast` where `point_spec` controls expected input shape (single point, array, tuple, leanest object, raw coords). Handy when benchmarking large batches of point comparisons.

```python
print(p2d(3,4).eq_fast(p2d(3,4), point_spec=1))
print(p2d(3,4).eq_fast([p2d(1,2), p2d(3,4)], point_spec=2))
print(p2d(3,4).eq_fast((p2d(1,4), p2d(3,4)), point_spec=2))
print(p2d(3,4).eq_fast(p2d_leanest(3,4), point_spec=3))
print(p2d(3,4).eq_fast([p2d_leanest(1,4), p2d_leanest(3, 4)], point_spec=4))
print(p2d(3,4).eq_fast([1,2], point_spec=5))
print(p2d(3,4).eq_fast([[1,2]], point_spec=6))
print(p2d(3,4).eq_fast([[1,2],[3,4],[5,6]], point_spec=7))
print(p2d(3,4).eq_fast([[3],[4],[5]], point_spec=8))
```

### Inequality via `!=` and `.ne()`
Covers both operator and method-based inequality checks to see how `Point2d` and `p2d_leanest` treat mismatched coordinates, sequences, and lean vs full objects.

```python
print(Point2d(3, 4) != p2d_leanest(3, 4))
print(Point2d(3, 4) != [p2d_leanest(1, 4), p2d_leanest(3, 4)])
print(Point2d(3, 4) != Point2d(3, 4))
print(Point2d(3, 4) != [Point2d(1, 2), Point2d(3, 4)])
print(Point2d(3, 4) != (Point2d(1, 4), Point2d(3, 4)))
print(Point2d(3, 4) != [[1, 2], [3, 4], [5, 6]])
print(Point2d(3, 4) != [[3, 4]])
print(Point2d(3, 4) != [[3], [4]])

print(Point2d(3, 4).ne(p2d_leanest(3, 4)))
print(Point2d(3, 4).ne([p2d_leanest(1, 4), p2d_leanest(3, 4)]))
print(Point2d(3, 4).ne(Point2d(3, 4)))
print(Point2d(3, 4).ne([Point2d(1, 2), Point2d(3, 4)]))
print(Point2d(3, 4).ne((Point2d(1, 4), Point2d(3, 4))))
print(Point2d(3, 4).ne([[1, 2], [3, 4], [5, 6]]))
print(Point2d(3, 4).ne([[3, 4]]))
print(Point2d(3, 4).ne([[3], [4]]))
```

### from_intersection_two_lines
Overloaded return.

Use `Point2d.from_intersection_two_lines` to classify line interactions (collinear overlap, disjoint collinear, and proper intersection). Covers both deterministic and ambiguous cases to show returned structures.

```python
from upxo.geoEntities.sline2d import Sline2d
# Example-1: Collinear lines Case 1
# ----------------------------------
la = Sline2d.by_coord([0, 0], [1, 1])
lb = Sline2d.by_coord([0.1, 0.1], [1.8, 1.8])
print(Point2d.from_intersection_two_lines(la, lb, tool='upxo'))

# Example-2: Collinear lines Case 2
# ----------------------------------
la = Sline2d.by_coord([0, 0], [1, 1])
lb = Sline2d.by_coord([0.1, 0.1], [0.8, 0.8])
print(Point2d.from_intersection_two_lines(la, lb, tool='upxo'))

# Example-3: Collinear lines Case 3
# ----------------------------------
la = Sline2d.by_coord([0, 0], [1, 1])
lb = Sline2d.by_coord([-0.1, -0.1], [1.8, 1.8])
print(Point2d.from_intersection_two_lines(la, lb, tool='upxo'))

# Example-4: Collinear lines Case 4
# ----------------------------------
la = Sline2d.by_coord([0, 0], [1, 1])
lb = Sline2d.by_coord([1.8, 1.8], [0.1, 0.1])
print(Point2d.from_intersection_two_lines(la, lb, tool='upxo'))

# Example-5: non-collinear lines Case 5
# ----------------------------------
la = Sline2d.by_coord([0, 0], [1, 1])
lb = Sline2d.by_coord([0, 1], [1, 0])
print(Point2d.from_intersection_two_lines(la, lb, tool='upxo'))

# Example-6: Collinear lines Case 6
# ----------------------------------
la = Sline2d.by_coord([0.1, 0.1], [0.8, 0.8])
lb = Sline2d.by_coord([0, 0], [1, 1])
print(Point2d.from_intersection_two_lines(la, lb, tool='upxo'))

# Example-7: Non-Collinear lines Case 7a
# ----------------------------------
la = Sline2d.by_coord([0, 0], [1, 1])
lb = Sline2d.by_coord([0, 0], [1, 0])
print(Point2d.from_intersection_two_lines(la, lb, tool='upxo'))

# Example-8: Non-Collinear lines Case 7b
# ----------------------------------
la = Sline2d.by_coord([0, 0], [1, 1])
lb = Sline2d.by_coord([-0,-0], [1, 0])
print(Point2d.from_intersection_two_lines(la, lb, tool='upxo'))

# Example-9: Collinear lines Case 8
# ----------------------------------
la = Sline2d.by_coord([0,0], [1,1])
lb = Sline2d.by_coord([0,0], [1,1])
print(Point2d.from_intersection_two_lines(la, lb, tool='upxo'))

# Example-10: Collinear lines Case 9
# ----------------------------------
la = Sline2d.by_coord([0,0], [1,1])
lb = Sline2d.by_coord([0,0], [-1,-1])
print(Point2d.from_intersection_two_lines(la, lb, tool='upxo'))

# Example-11: Collinear lines Case 10
# ----------------------------------
la = Sline2d.by_coord([0,0], [1,0])
lb = Sline2d.by_coord([0,1], [1,1])
print(Point2d.from_intersection_two_lines(la, lb, tool='upxo'))
```

### rotate_about_point
Rotate self point about another point.

Demonstrates rotating a point about another point using degrees, including negative angles and mixed input types (lean, tuple, nested lists). Shows effect of `update`/`throw`.

```python
dut = {'degree': True, 'update': True, 'throw': True}

print(p2d(1, 0).rotate_about_point(p2d(0, 0), -45, **dut))
print(p2d(1, 0).rotate_about_point(p2d(0, 0), +45, **dut))
print(p2d(0, 0).rotate_about_point(p2d(-1, 0), 45, **dut))
print(p2d(0, 0).rotate_about_point(p2d(-1, 0), -45, **dut))
print(p2d(0, 0).rotate_about_point(p2d(-1, 0), 45, **dut))
print(p2d(0.70710678,0.70710678).rotate_about_point(p2d(0, 0), 45, **dut))
print(p2d(0.70710678,0.70710678).rotate_about_point(p2d(0, 0), -45, **dut))
print(p2d(-0.70710678,0.70710678).rotate_about_point(p2d(0, 0), -45, **dut))

print(p2d(1, 0).rotate_about_point((0, 0), 45, **dut))
print(p2d(1, 0).rotate_about_point(([0], [0]), 45, **dut))

print(p2d(0, 0).rotate_about_point(p2d(0, 0), -45, **dut))
```

### array_by_clustering
#### Explanation
Make an array of points with self as centroid (approx).
#### Parameters

**n**: number of points to create.
**r**: radius
**distribution**: Specifies the distribution type. Options include:
- urand - Uniform random
- nrand - random uniform
- qrand - quadratic random
- crand - cubic random

**return_type**: Type of the return object. Options include:
- coord_list_2d - [[x0, ..., xn], [y0, ..., yn]]
- coords_2d - [[x0, y0], ..., [xn, yn]]
- upxo_2d - upxo 2D point object
- coord_list_3d - [[x0, ..., xn], [y0, ..., yn], [z0, ..., zn]]
- coords_3d - [[x0, y0, z0], ..., [xn, yn, zn]]
- upxo_3d - upxo 3D point object
- gmsh - GMSH poit object
- pyvista - Py-Vista object
**zloc**: Value of z-coordinate if return type is 3D.

#### Return
LIst of point coordinates.

#### Comments
When random, greater the value of n, nearer would be the self to
the centroid.

Executes `array_by_clustering` to generate synthetic neighborhoods around `Point2d`, swapping return types (raw coords, lean points, shapely, gmsh, pyvista) to compare interoperability paths.

```python
print(p2d(0,0).array_by_clustering(n=10, r=1))
print(p2d(0,0).array_by_clustering(n=10, r=1, return_type='coords_2d'))
print(p2d(0,0).array_by_clustering(n=10, r=1, return_type='upxo_2d'))
print(p2d(0,0).array_by_clustering(n=10, r=1, return_type='upxo_2d_leanest'))
print(p2d(0,0).array_by_clustering(n=10, r=1, return_type='shapely'))
print(p2d(0,0).array_by_clustering(n=10, r=1, return_type='gmsh'))
print(p2d(0,0).array_by_clustering(n=10, r=1, return_type='pyvista'))
print(p2d(0,0).array_by_clustering(n=10, r=1, return_type='coords_3d'))
print(p2d(0,0).array_by_clustering(n=10, r=1, return_type='coord_list_3d'))
print(p2d(0,0).array_by_clustering(n=10, r=1, return_type='mulpoint2d'))
```
