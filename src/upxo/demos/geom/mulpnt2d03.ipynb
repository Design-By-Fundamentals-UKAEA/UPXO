# Generate 2D multi-point object: feed in a set of points

## Overview

This notebook demonstrates the creation and manipulation of 2D multi-point objects using the `mulpoint2d` class from the `upxo.geoEntities` module. The `mulpoint2d` class provides flexible methods for initializing collections of 2D points from various input formats.

## Purpose

The primary goal is to showcase different initialization methods for creating multi-point geometric entities, which is useful for:
- Geospatial data processing
- Computational geometry applications
- Visualization of point datasets
- Data structure conversion and standardization

## Key Components

### Required Modules
- `mulpoint2d`: Main class for handling multiple 2D points
- `Point2d (p2d)`: Individual 2D point representation
- `numpy`: For array operations and data manipulation

### Initialization Methods

The `mulpoint2d` class supports multiple initialization strategies to accommodate different data formats:

#### **method options**
1. **`up2d`**: Initialize from Point2d objects
    - Parameter: `point_objects` - list of Point2d instances
    - Use case: When working with existing Point2d objects
    - *Dev note*: Look up `dth.opt.upxo_point2d` for more details.

2. **`xcoord_ycoord`**: Initialize from separate x and y coordinate lists
    - Parameters: `coordx` and `coordy` - separate lists for x and y coordinates
    - Use case: When coordinates are stored in separate arrays
    - *Dev note*: Look up `dth.opt.coord_xlist_ylist` for more details.

3. **`xy_coord_pairs_list`**: Initialize from coordinate pair lists
    - Parameter: `coordxy` - list of [x, y] pairs
    - Use case: When data is in row-wise coordinate pair format
    - *Dev note*: Look up `dth.opt.coord_pairs_point2d_list` for more details.

4. **`xylist` or `coord_list`**: Initialize from transposed coordinate array
    - Parameter: `coordxy` - numpy array with shape (2, n) where first row is x-coords and second is y-coords
    - Use case: When working with column-major coordinate arrays
    - *Dev note*: Look up `dth.opt.coord_point2d_list` for more details.

5. **`mp2d`**: Initialize from existing multi-point objects
    - Parameter: `mulpoint_objects` - existing mulpoint2d instances
    - Use case: For copying or combining existing multi-point objects
    - *Dev note*: Look up `dth.opt.upxo_mp` for more details.

6. **Grid generation methods** (`trigrid1`, `trigrid2`, `recgrid`, `hexgrid`, `random`): Generate points in specific patterns
    - Parameter: `a` - grid parameters
    - Use case: For generating structured or random point distributions
    - *Dev note*: Look up `dth.opt.upxo_mp_grid_type` for more details.

### Additional Parameters

- **`lean`**: Controls the data storage optimization level (e.g., `'leanest'` for minimal memory footprint)

## Workflow

Each section below demonstrates a specific initialization method with practical examples, allowing you to choose the most appropriate method for your data format and use case.

```python
from upxo.geoEntities.mulpoint2d_old import mulpoint2d
from upxo.geoEntities.point2d import Point2d as p2d
import numpy as np
```

# Option: `up2d`

```python
points = [p2d(1, -4), p2d(3, 4), p2d(-3, -4)]
```

```python
points
```

```python
sp = mulpoint2d(method='up2d', point_objects=points, lean='leanest')
```

# Option: `xcoord_ycoord`

```python
x = [1, 2, 3, 4, 5]
y = [9, 5, 7, 4, 3]
sp = mulpoint2d(method='xcoord_ycoord', coordx=x, coordy=y, lean='leanest')
sp.points
```

# Option: `xy_coord_pairs_list`

```python
xy = [[1, 9], [2, 5], [3, 7], [4, 4], [5, 3]]
# xy = [[1, 2, 3, 4, 5], [9, 5, 7, 4, 3]]
# coordxy = np.vstack((x, y)).T
sp = mulpoint2d(method='xy_coord_pairs_list', coordxy=xy, lean='leanest')
```

# Option: `xylist`

```python
xy = [[1, 9], [2, 5], [3, 7], [4, 4], [5, 3]]
coordxy = np.array(xy).T
sp = mulpoint2d(method='xylist', coordxy=coordxy, lean='leanest')
```
