# Generate 2D multi-point object: n number of uniformly distributed random points

This notebook demonstrates the creation and visualization of a 2D multi-point geometric entity using the `mulpoint2d` class from the UPXO library.

## Overview

The `mulpoint2d` class generates a collection of points in 2D space based on specified parameters. This implementation creates a set of randomly distributed points within defined boundaries using uniform sampling techniques.

## Configuration Parameters

- **method**: `'random'` - Specifies the point generation method
- **gridding_technique**: `'random'` - Defines how points are spatially distributed
- **sampling_technique**: `'uniform'` - Ensures uniform distribution of points
- **nrndpnts**: `8` - Number of random points to generate
- **randuni_calc**: `'uniform'` - Random number generation approach
- **space**: `'linear'` - Spatial coordinate system
- **xbound**: `[0, 100]` - X-axis boundary limits
- **ybound**: `[0, 50]` - Y-axis boundary limits
- **lean**: `'leanest'` - Optimization parameter for point distribution
- **make_point_objects**: `True` - Creates individual point objects
- **make_ckdtree**: `True` - Generates a k-d tree for spatial queries
- **vis**: `False` - Visualization during creation disabled
- **print_summary**: `True` - Prints summary information upon creation

## Output

The generated multi-point object contains 8 uniformly distributed points with a centroid located at coordinates (57.49, 26.45).

```python
from upxo.geoEntities.mulpoint2d_old import mulpoint2d
```

```python
sp = mulpoint2d(method='random',
                gridding_technique='random',
                sampling_technique='uniform',
                nrndpnts=8,
                randuni_calc='uniform',
                space='linear', xbound=[0, 100],
                ybound=[0, 50], lean='leanest',
                make_point_objects=True,
                make_ckdtree=True,
                vis=False, print_summary=True)
```

```python
sp.plot()
```
